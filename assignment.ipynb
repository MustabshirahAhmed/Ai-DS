```python
#using while loop

# Keep asking for the correct bank login password until entered correctly
correct_password = "bank123"
entered_password = input("Enter your bank login password: ")
while entered_password != correct_password:
    print("Incorrect password. Please try again.")
    entered_password = input("Enter your bank login password: ")
print("Login successful!")

# Conditional transaction counter
transaction_counter = 1

while transaction_counter <= 10:
    action = input("Enter transaction type (credit/debit/exit): ").lower()
    if action == "credit":
        amount = float(input("Enter credit amount: "))
        balance += amount
        print(f"Transaction {transaction_counter}: CREDIT of {amount}. New balance: {balance}")
        transaction_counter += 1
    elif action == "debit":
        amount = float(input("Enter debit amount: "))
        balance -= amount
        print(f"Transaction {transaction_counter}: DEBIT of {amount}. New balance: {balance}")
        transaction_counter += 1
    elif action == "exit":
        print("Exiting transaction loop.")
        break
    else:
        print("Invalid transaction type. Please enter 'credit', 'debit', or 'exit'.")
```
